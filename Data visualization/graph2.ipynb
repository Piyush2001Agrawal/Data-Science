```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("iris")
df
```

```python
sns.pairplot(df, hue='species', palette='Set2')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.boxplot(data=df, x='species', y='petal_length', palette='Set3')
plt.title('Petal Length by Species')
plt.show()
```

```python
sns.boxplot(data=df, x='species', y='petal_width', palette='YlGnBu')
plt.title('Petal Width by Species')

plt.show()
```

```python
g = sns.FacetGrid(df, col='species', palette='Set2')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
plt.show()
```
