```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('titanic')
df
```

```python
sns.countplot(df , y='survived')
plt.title('Count of Survived vs Not Survived')
plt.show()
```

```python
sns.countplot(df , x='survived')
plt.title('Count of Survived vs Not Survived')
plt.show()
```

```python
sns.countplot(df , x='survived', hue= 'sex', palette='PuBu')
plt.title('Count of Survived vs Not Survived by Gender')
plt.show()
```

```python
sns.countplot(df , x='survived', hue= 'pclass')
plt.title('Count of Survived vs Not Survived by Gender')
plt.show()
```

```python
sns.barplot(df, x= 'class', y ='survived', palette='PuBu')
plt.title('Survival Rate by Class') 
plt.show()
```

```python
sns.histplot(df, x='age', bins = 30, kde = True, color='purple')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
sns.boxplot(data=df, x='class', y='age', palette='Set2')
plt.title('Age Distribution by Class')
```

```python
sns.violinplot( data=df,x='survived', y='age', hue='sex', split=True)
plt.title('Age Distribution by Gender and Survival')
```

```python
df['survived'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Survival Rate')
plt.show()
```

```python
sns.boxplot(data=df, x='pclass', y='age', palette='Set2')
plt.title('Age Distribution by Class')
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='survived')
plt.title('Fare vs Age by Survival')
```

```python
g = sns.FacetGrid(df, col='survived', row='pclass', margin_titles=True,palette='PuBu')
g.map(sns.histplot, 'age')
```

```python
sns.countplot(data=df, x='embark_town', palette='magma')
plt.title('Passenger Count by Embarkation Point')
```

```python
sns.kdeplot(data=df, x='age', hue='survived', fill=True, palette='Reds')
plt.title('Age Distribution by Survival')
```

```python
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='YlOrRd')
plt.title('Pairplot of Age, Fare, Class and Survival')
```

```python
sns.stripplot(data=df, x='pclass', y='age', hue='survived', jitter=True, dodge=True, palette='YlGnBu')
plt.title('Age Distribution by Class and Survival')
```

```python
sns.boxenplot(data=df, x='embarked', y='fare', palette='plasma')
plt.title('Fare Distribution by Embarkation Point')
```

```python
sns.jointplot(data=df, x='age', y='fare', hue='survived', kind='scatter', palette='magma')
plt.suptitle('Age vs Fare by Survival')
```

```python
sns.countplot(data=df, x='embark_town', hue='survived', palette='magma')
plt.title('Passenger Count by Embarkation Point and Survival')
```

```python
sns.lineplot(data=df, x='embarked', y='fare', hue='pclass', ci=None, palette='coolwarm')
plt.title('Fare Distribution by Embarkation Point and Class')
```

```python
sns.catplot(data=df, x='pclass', hue='survived', col='sex', kind='count', palette='viridis')
```

# Plotly Graph

```python
import plotly.express as px
```

```python
px.scatter(df,x='age',y='fare',color='survived',title='Age vs Fare by Survival',template='plotly_dark')
```

```python
px.sunburst(df,path=['pclass','sex','survived'],values='age',
            title='Survival by class and gender', width=500, height=500,
        template='plotly_dark', color='age', color_continuous_scale='RdBu')
```

```python
px.histogram(df,x='age',nbins=30,color='survived',title='Age Distribution by Survival',
           template='plotly_dark',)
```

```python
px.treemap(df,path=['pclass','survived'],values='age',
          title='Survival by Class', width=800,height=800,
          template='plotly_dark',
          color='age', color_continuous_scale='plasma')
```

```python
survived_counts= df['survived'].value_counts()
px.pie(names=survived_counts.index, values=survived_counts.values,
       title='Survival Rate', template='plotly_dark',
       width=500, height=500)
```

```python
px.box(df,x='pclass',y='fare',color='pclass',
       title='Fare Distribution by Class', template='plotly_dark',
       width=500, height=500)
```

```python
px.violin(df, x='survived', y='age', color='sex', 
                 box=True, points='all', 
                 title='Age Distribution by Survival and Gender', 
                 template='plotly_dark')
```

```python
px.bar(df, x='embark_town', y='survived',color='embark_town', 
              title='Survival Rate by Embarkation Town', 
              )
```

```python
px.scatter_matrix(df, dimensions=['age', 'fare', 'pclass'], color='survived',
                          title='Pair Plot of Age, Fare, Class, and Survival', 
                          color_continuous_scale='viridis',)
```

```python
px.line(df, x='embark_town', y='fare', color='pclass', 
               title='Fare by Embarkation Point and Class', 
               )
```

```python
px.density_contour(df, x='age', y='fare', color='survived',
                          title='Density Contour of Age vs Fare by Survival',
                          )
```

```python
embark_survival = df.groupby('embark_town')['survived'].mean().reset_index()
px.line_polar(embark_survival, r='survived', theta='embark_town',
                     line_close=True, title='Survival Rate by Embarkation Town',
                     template='plotly_dark')
```

```python
px.histogram(df, x='fare', color='pclass', nbins=20,
                    title='Fare Distribution by Class')
```

```python

px.scatter(df, x='age', y='fare', size='sibsp', color='survived',
                  title='Bubble Chart: Age vs Fare with SibSp',
                  template='plotly_dark')
```

```python
px.scatter_3d(df, x='age', y='fare', z='pclass', color='survived',
                      title='3D Scatter Plot: Age, Fare, and Survival',color_continuous_scale='viridis'
                )
```
