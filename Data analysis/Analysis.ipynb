```python
# loading data for missing value handling
df=  sns.load_dataset("titanic")
df
```

```python
df.isnull().sum()
```

```python
avg_ag = df['age'].mean()
print(avg_ag)
```

```python
avg_ag = round(df['age'].mean())
print(avg_ag)
```

```python
df['age'].fillna(avg_ag, inplace=True)
df.isnull().sum()
```

```python
# Deleteling the rows with missing values
df.dropna(subset=['embarked'], inplace=True)
df.isnull().sum()
```
